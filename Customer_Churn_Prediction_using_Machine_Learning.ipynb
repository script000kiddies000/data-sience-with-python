<div id="theorycontent" ><h2>Menghapus Unnecessary Columns dari data</h2><div class="cell text_cell unselected rendered" tabindex="2">

<p>Selanjutnya kita akan mengapus kolom yang tidak akan diikutsertakan dalam pemodelan, kemudian simpan dengan nama&nbsp;<code>cleaned_df</code>. Tampilkan 5 rows teratas nya.</p>

&nbsp;</div>
Hasil yang akan ditampilkan adalah sebagai berikut.</div>
</div>
</div></div>

```python
#Remove the unnecessary columns customerID & UpdatedAt
cleaned_df = df_load.drop(['customerID','UpdatedAt'], axis=1)
cleaned_df.head()
```

<div id="theorycontent"><h2>Encoding Data</h2><p>Gunakan data dari hasil dan analisa sebelumnya&nbsp;<code>cleaned_df</code>, untuk merubah value dari data yang masih berbentuk string untuk diubah ke dalam bentuk numeric menggunakan&nbsp;<code>LabelEncoder()</code>. Gunakan&nbsp;<code>describe()</code>&nbsp;untuk melihat deskripsi datanya.</p>
<p>Barikut adalah hasilnya :</p>
<p></p></div>

```python
from sklearn.preprocessing import LabelEncoder
#Convert all the non-numeric columns to numerical data types
for column in cleaned_df.columns:
    if cleaned_df[column].dtype == np.number: continue
# Perform encoding for each non-numeric column
    cleaned_df[column] = LabelEncoder().fit_transform(cleaned_df[column])
print(cleaned_df.describe())
```

<div id="theorycontent"><h2>Splitting Dataset</h2><p>Gunakan data dari hasil dan analisa sebelumnya&nbsp;<code>cleaned_df</code>, untuk dibagi datasetnya menjadi 2 bagian (70% training &amp; 30% testing) berdasarkan variable predictor (X) dan targetnya (Y). Gunakan train_test_split() untuk membagi data tersebut. Sertakan&nbsp;<code>value_counts</code>&nbsp;untuk mengecek apakah pembagian sudah sama proporsinya. Simpan hasil spliting data menjadi&nbsp;<code>x_train</code>,&nbsp;<code>y_train</code>,&nbsp;<code>x_test</code>&nbsp;&amp;&nbsp;<code>y_test</code></p>
<p><code>Hasil yang diharapkan adalah sebagai berikut :</code></p>
<p><code></code></p></div>

```python
from sklearn.model_selection import train_test_split
# Predictor dan target
X = cleaned_df.drop('Churn', axis = 1)
y = cleaned_df['Churn']
# Splitting train and test
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Print according to the expected result
print('Jumlah baris dan kolom dari x_train adalah:', x_train.shape,', sedangkan Jumlah baris dan kolom dari y_train adalah:', y_train.shape)
print('Prosentase Churn di data Training adalah:')
print(y_train.value_counts(normalize=True))
print('Jumlah baris dan kolom dari x_test adalah:', x_test.shape,', sedangkan Jumlah baris dan kolom dari y_test adalah:', y_test.shape)
print('Prosentase Churn di data Testing adalah:')
print(y_test.value_counts(normalize=True))
```

<div id="theory"><h2>Kesimpulan</h2><p>Setelah kita analisis lebih lanjut, ternyata ada kolom yang tidak dibutuhkuan dalam model, yaitu Id Number pelanggannya (<code>customerID</code>) &amp; periode pengambilan datanya (<code>UpdatedAt</code>), maka hal ini perlu dihapus. Kemudian kita lanjut mengubah value dari data yang masih berbentuk string menjadi numeric melalui encoding, setelah dilakukan terlihat di persebaran datanya khususnya kolom min dan max dari masing masing variable sudah berubah menjadi 0 &amp; 1. Tahap terakhir adalah membagi data menjadi 2 bagian untuk keperluan modelling, setelah dilakukan terlihat dari jumlah baris dan kolom masing-masing data sudah sesuai &amp; prosentase kolom churn juga sama dengan data di awal, hal ini mengindikasikan bahwasannya data terpisah dengan baik dan benar.</p></div>

<div id="theorycontent"><h2>Pembuatan Model</h2><p>Selanjutnya kita akan membuat model dengan menggunakan Algoritma Logistic Regression.</p>

Gunakan <code>LogisticRegression()</code>&nbsp;memanggil algoritma tersebut, fit ke data train dan simpan sebagai&nbsp;<code>log_model</code></div>
&nbsp;</div>
Output dari codingan pada tahap ini adalah sebagai berikut</div>
&nbsp;</div>
</div>
</div></div>

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(solver='lbfgs', max_iter=400).fit(x_train, y_train)
print('Model Logistic Regression yang terbentuk adalah: \n',log_model)
```

<div id="theorycontent"><h2>Performansi Model Training - Menampilkan Metrics </h2><p>Setelah kita membuat modelnya, maka lakukan perhitungan untuk memperoleh classification reportnya dan confusion matrixnya di data training seperti hasil di bawah ini. Gunakan&nbsp;<code>classification_report()</code>&nbsp;&amp;&nbsp;<code>confusion_matrix()</code>.</p>
<p>Output pada tahap ini adalah sebagai berikut.</p>
<p></p>
<p>&nbsp;</p></div>

```python


from sklearn.metrics import classification_report
# Predict
y_train_pred = log_model.predict(x_train)
# Print classification report
print('Classification Report Training Model (Logistic Regression) :')
print(classification_report(y_train, y_train_pred))
```

<div id="theorycontent" ><h2>Performansi Model Training - Menampilkan Plots</h2><p>Setelah mendapatkan hasil classification report pada tahap sebelumnya, sekarang kita akan melakukan visualisasi terhadap report tersebut.&nbsp;</p>
<p></p></div>

```python
from sklearn.metrics import confusion_matrix
from matplotlib import pyplot as plt
import seaborn as sns

# Form confusion matrix as a DataFrame
confusion_matrix_df = pd.DataFrame((confusion_matrix(y_train, y_train_pred)), ('No churn', 'Churn'), ('No churn', 'Churn'))

# Plot confusion matrix
plt.figure()
heatmap = sns.heatmap(confusion_matrix_df, annot=True, annot_kws={'size': 14}, fmt='d', cmap='YlGnBu')
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)

plt.title('Confusion Matrix for Training Model\n(Logistic Regression)', fontsize=18, color='darkblue')
plt.ylabel('True label', fontsize=14)
plt.xlabel('Predicted label', fontsize=14)
plt.show()
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Menentukan Algoritma Model Terbaik</h2><p>Model yang baik adalah model yang mampu memberikan performa bagus di fase training dan testing model.</p>
<ul>
<li><code>Over-Fitting</code>&nbsp;adalah suatu kondisi dimana model mampu memprediksi dengan sangat baik di fase training, akan tetapi tidak mampu memprediksi sama baiknya di fase testing.</li>
<li><code>Under-Fitting</code>&nbsp;adalah suatu kondisi dimana model kurang mampu memprediksi dengan baik di fase training, akan tetapi mampu memprediksi dengan baik di fase testing.</li>
<li><code>Appropriate-Fitting</code>&nbsp;adalah suatu kondisi dimana model mampu memprediksi dengan baik di fase training maupun di fase testing.</li>
</ul>
